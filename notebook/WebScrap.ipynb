```python
#@title 🔧 Step 1 — Setup { display-mode: "form" }
#@markdown Install Chromium, Selenium, and dependencies. Takes ~60s on first run.
#@markdown ---

import os, sys, subprocess, shutil, warnings
warnings.filterwarnings("ignore", category=UserWarning)
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"

# ── Clone repo (src/ modules live here) ──
REPO_DIR = "/content/Z-Web-Scraper"
if not os.path.isdir(os.path.join(REPO_DIR, "src")):
    subprocess.run(["git", "clone", "-q", "https://github.com/festverse/Z-Web-Scraper.git", REPO_DIR], capture_output=True)
else:
    subprocess.run(["git", "-C", REPO_DIR, "pull", "-q"], capture_output=True)
if REPO_DIR not in sys.path:
    sys.path.insert(0, REPO_DIR)

# ── Install system packages ──
subprocess.run(["apt-get", "-y", "-qq", "update"], capture_output=True)
subprocess.run(["apt-get", "-y", "-qq", "install", "-qq",
    "chromium-browser", "chromium-chromedriver",
    "xvfb", "fonts-liberation", "libnss3", "libatk-bridge2.0-0",
    "libdrm2", "libxcomposite1", "libxdamage1", "libxrandr2",
    "libgbm1", "libasound2", "libpangocairo-1.0-0", "libgtk-3-0"],
    capture_output=True)
subprocess.run(["cp", "/usr/lib/chromium-browser/chromedriver", "/usr/bin"], capture_output=True)

# ── Install Python packages ──
subprocess.run([sys.executable, "-m", "pip", "install", "-q",
    "selenium", "beautifulsoup4", "requests", "lxml"], capture_output=True)

# ── Import UI from repo ──
from src.ui import show_header, show_ok, show_warn, show_err, show_info, show_step

show_header("🔧", "Setup", "Installing Chromium, Selenium & dependencies")

# ── Detect Chromium ──
show_step("Detecting Chromium...")
CHROME_BIN = None
for p in ["/usr/bin/chromium-browser", "/usr/bin/chromium", "/usr/bin/google-chrome",
          shutil.which("chromium-browser"), shutil.which("chromium"), shutil.which("google-chrome")]:
    if p and os.path.isfile(p): CHROME_BIN = p; break
if CHROME_BIN:
    ver = subprocess.run([CHROME_BIN, "--version"], capture_output=True, text=True)
    show_ok(f"Chromium: {ver.stdout.strip() if ver.returncode == 0 else CHROME_BIN}")
else:
    show_err("Chromium not found — restart runtime")

# ── Selenium test ──
show_step("Testing Selenium...")
try:
    from src.scraper import _build_driver
    _d = _build_driver(CHROME_BIN)
    _d.get("data:text/html,<h1>OK</h1>")
    assert "OK" in _d.page_source
    _d.quit()
    show_ok("Selenium verified")
except Exception as e:
    show_err(f"Selenium failed: {str(e)[:200]}")

# ── Save config ──
import json
with open("/tmp/Z-Web-Scraper_driver_config.json", "w") as f:
    json.dump({"chrome_bin": CHROME_BIN}, f)

OUTPUT_DIR = "/content/Z-Web-Scraper_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)
show_ok(f"Output: {OUTPUT_DIR}")
show_ok("Setup complete!", "🎉")
show_info("Run the next cell to scrape a website", "➡️")
```

```python
#@title 🕷️ Step 2 — Scrape Website { display-mode: "form" }
#@markdown ### Enter URL
#@markdown Paste any website URL below.
url = "" #@param {type:"string"}
#@markdown ### Settings
timeout = 30 #@param {type:"slider", min:10, max:300, step:5}
wait_after_load = 3 #@param {type:"slider", min:0, max:60, step:1}
auto_scroll = True #@param {type:"boolean"}
custom_js = "" #@param {type:"string"}
#@markdown ### Output
save_html_file = True #@param {type:"boolean"}
show_preview = True #@param {type:"boolean"}
show_links = True #@param {type:"boolean"}
#@markdown ---

import os, sys, json, html as html_lib
from datetime import datetime, timezone
from urllib.parse import urlparse

REPO_DIR = "/content/Z-Web-Scraper"
if REPO_DIR not in sys.path: sys.path.insert(0, REPO_DIR)

# ── Import everything from repo ──
from src.scraper import scrape_url, save_html, save_metadata
from src.ui import (
    show_header, show_ok, show_warn, show_err, show_info, show_step,
    show_stats, show_html_preview, show_links_table, show_meta_tags,
)

# ── Load Chrome binary ──
import shutil, hashlib
CHROME_BIN = None
try:
    with open("/tmp/Z-Web-Scraper_driver_config.json") as f:
        CHROME_BIN = json.load(f).get("chrome_bin")
except FileNotFoundError: pass
if not CHROME_BIN:
    for p in ["/usr/bin/chromium-browser", "/usr/bin/chromium", shutil.which("chromium-browser"), shutil.which("chromium")]:
        if p and os.path.isfile(p): CHROME_BIN = p; break
if not CHROME_BIN:
    show_err("Chromium not found. Run Step 1."); raise SystemExit()

# ── Validate URL ──
if not url or not url.strip():
    show_err("Enter a URL above"); raise SystemExit()
url = url.strip()
if not url.startswith(('http://', 'https://')): url = 'https://' + url
parsed = urlparse(url)
if not parsed.netloc:
    show_err(f"Invalid URL: {url}"); raise SystemExit()

show_header("🕷️", "Scraping", url)

# ── Progress callback ──
def on_progress(event, msg):
    if event == "ok": show_ok(msg)
    elif event == "step": show_step(msg)
    elif event == "warn": show_warn(msg)
    elif event == "err": show_err(msg)

# ── Scrape using repo module ──
OUTPUT_DIR = "/content/Z-Web-Scraper_output"
result = scrape_url(
    url=url,
    timeout=timeout,
    wait_after=wait_after_load,
    scroll=auto_scroll,
    execute_js=custom_js if custom_js.strip() else None,
    chrome_bin=CHROME_BIN,
    callback=on_progress,
)

# ── Display results ──
show_stats([
    ("📄", "Characters", f"{result['char_count']:,}"),
    ("📦", "Size", f"{result['file_size_bytes'] / 1024:.1f} KB"),
    ("🔗", "Links", str(result['link_count'])),
    ("🖼️", "Images", str(result['image_count'])),
    ("⏱️", "Load Time", f"{result['load_time_s']}s"),
    ("📜", "Scrolls", str(result['scroll_count'])),
])

if result['final_url'] != url:
    show_info(f"Redirected: {url} → {result['final_url']}", "↪️")
if result['title']:
    show_ok(f"Title: {result['title']}", "🏷️")
desc = result['meta_tags'].get('og:description') or result['meta_tags'].get('description')
if desc: show_info(f"Description: {desc[:200]}")

# ── Save files ──
if save_html_file:
    domain = parsed.netloc.replace("www.", "")
    safe_name = "".join(c if c.isalnum() or c in ('-', '.') else '_' for c in domain)
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    filepath = os.path.join(OUTPUT_DIR, f"{safe_name}_{ts}.html")
    save_html(result['html'], filepath)
    show_ok(f"Saved: {filepath}", "💾")
    meta_path = filepath.replace('.html', '_meta.json')
    save_metadata(result, meta_path)
    show_ok(f"Metadata: {meta_path}", "📋")

if show_preview: show_html_preview(result['html'])
if show_links: show_links_table(result['links'])
show_meta_tags(result['meta_tags'])

show_header("✅", "Done!", f"Scraped {result['title'] or url} in {result['load_time_s']}s")
show_info("Run again with a different URL to scrape more pages", "🔄")
```

```python
#@title 💾 Step 3 — Download & Export { display-mode: "form" }
#@markdown Download all scraped files as a ZIP archive.
#@markdown ---

import os, sys, glob, zipfile
from datetime import datetime

REPO_DIR = "/content/Z-Web-Scraper"
if REPO_DIR not in sys.path: sys.path.insert(0, REPO_DIR)

from src.ui import show_header, show_ok, show_warn, show_info, show_step

OUTPUT_DIR = "/content/Z-Web-Scraper_output"
show_header("💾", "Export", "Download scraped files")

html_files = sorted(glob.glob(os.path.join(OUTPUT_DIR, "*.html")))
json_files = sorted(glob.glob(os.path.join(OUTPUT_DIR, "*.json")))
all_files = html_files + json_files

if not all_files:
    show_warn("No files found. Run Step 2 first.")
else:
    show_info(f"Found {len(html_files)} HTML + {len(json_files)} metadata files")
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    zip_path = f"/content/Z-Web-Scraper_export_{ts}.zip"
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zf:
        for f in all_files: zf.write(f, os.path.basename(f))
    show_ok(f"Created {os.path.basename(zip_path)} ({os.path.getsize(zip_path) / 1024:.1f} KB)")
    for f in all_files:
        show_info(f"  📄 {os.path.basename(f)} ({os.path.getsize(f) / 1024:.1f} KB)")
    try:
        from google.colab import files as colab_files
        show_step("Downloading...")
        colab_files.download(zip_path)
        show_ok(f"Downloaded {os.path.basename(zip_path)}")
    except ImportError:
        show_info(f"Not on Colab. File: {zip_path}")
    except Exception as e:
        show_warn(f"Download failed: {str(e)[:100]}")

show_header("✅", "All Done!", "Your scraped data is ready")
```
